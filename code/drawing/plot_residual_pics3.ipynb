```python
def lighten_color(color, amount=0.5):
    try:
        c = mc.cnames[color]
    except:
        c = color
    c = colorsys.rgb_to_hls(*mc.to_rgb(c))
    return colorsys.hls_to_rgb(c[0], 1 - amount * (1 - c[1]), c[2])

val = 0.6
# colors = {'HDA':lighten_color('blueviolet', val),'CI':lighten_color('skyblue', 1.0),'MinSum':lighten_color('darkseagreen', 1),
#           'BPD':lighten_color('k', val),'CoreHD':lighten_color('peru', 0.6),'GND':lighten_color('blue', 0.5),
#           'RatioCut':lighten_color('blue', 0.5),'FINDER':lighten_color('red', 1.0)}
colors = {'HDA':lighten_color('peru', 0.6),'FINDER':lighten_color('darkseagreen', 1),'ND-GRL':lighten_color('red', 1.0)}
```

```python
g = nx.read_edgelist('./g_test_7.txt')
degrees = nx.degree(g)
nodes = list(range(nx.number_of_nodes(g)))
edges = nx.number_of_edges(g)
# print(degrees['0'])
print(nx.degree_histogram(g))

nodes_weight_d = {}
for node in nodes:
    nodes_weight_d[str(node)] = 100*degrees[str(node)] / (2*edges) # 100只是观看方便，计算总weight会消去
print(len(nodes_weight_d.items()))
print(min(nodes_weight_d.values()),max(nodes_weight_d.values()),len(nx.degree_histogram(g)))


lines = 0
nodes_weight = []
SDNDRL_sol=[]
SDNDRL_anc=[]
for line in open('./FINDER_GA_g_test_7.txt', 'r'):
    if lines == 0:
        weight = np.double(line.strip())
        SDNDRL_anc.append(weight)
    else:
        weight = line.strip().split(',')
        weight[0] = int(weight[0])
        weight[1] = weight[1].strip(' ')
        weight[1] = np.double(weight[1])
        SDNDRL_sol.append(weight[0])
        SDNDRL_anc.append(weight[1])
#         if weight[1] >= 0.5:
#             SDNDRL_sol.append(weight[0])
#             SDNDRL_anc.append(weight[1])
#         else:
#             break
    nodes_weight.append(weight)
    lines = lines + 1
# print(nodes_weight)
# print(SDNDRL_sol,len(SDNDRL_sol))
# print(SDNDRL_anc,len(SDNDRL_anc))
# total_nodes_weight = np.sum(nodes_weight)
lines1 = 0
nodes_weight1 = []
FINDER_sol=[]
FINDER_anc=[]
for line in open('./FINDER_original_g_test_7.txt', 'r'):
    if lines1 == 0:
        weight = np.double(line.strip())
        FINDER_anc.append(weight)
    else:
        weight = line.strip().split(',')
        weight[0] = int(weight[0])
        weight[1] = weight[1].strip(' ')
        weight[1] = np.double(weight[1])
        FINDER_sol.append(weight[0])
        FINDER_anc.append(weight[1])
#         if weight[1] >= 0.5:
#             FINDER_sol.append(weight[0])
#             FINDER_anc.append(weight[1])
#         else:
#             break
    nodes_weight1.append(weight)
    lines1 = lines1 + 1
    
    
lines2 = 0
nodes_weight2 = []
HDA_sol=[]
HDA_anc=[]
for line in open('./HDA_g_test_7.txt', 'r'):
    if lines2 == 0:
        weight = np.double(line.strip())
        HDA_anc.append(weight)
    else:
        weight = line.strip().split(',')
        weight[0] = int(weight[0])
        weight[1] = weight[1].strip(' ')
        weight[1] = np.double(weight[1])
        HDA_sol.append(weight[0])
        HDA_anc.append(weight[1])
#         if weight[1] >= 0.5:
#             HDA_sol.append(weight[0])
#             HDA_anc.append(weight[1])
#         else:
#             break
    nodes_weight2.append(weight)
    lines2 = lines2 + 1
```

```python
total_nodes_weight = 0.0
# nodes_weight = []

prob_type = 'ND'



num_nodes = 2375 # from Yeast data
# num_nodes = 379 
# num_nodes = 198 
# num_nodes = 1133 
# num_nodes = 297 


# g = nx.read_edgelist('../data/crime.txt')
# degrees = nx.degree(g)
# nodes = list(range(nx.number_of_nodes(g)))

# for line in open('../data/%s/crime_weight.txt'%(prob_type), 'r'):
#     weight = np.double(line)
#     nodes_weight.append(weight)
for k,v in nodes_weight_d.items():
    total_nodes_weight = total_nodes_weight + v
# total_nodes_weight = np.sum(nodes_weight)

if prob_type == 'ND':
#     colors = {'MinSum':'green','GND':'blue','FINDER':'red'}
    markers = {'HDA':'s','FINDER':'H','ND-GRL':'D'} 
    methods = ['HDA','FINDER','ND-GRL']
    sol = {'HDA':[],'FINDER':[],'ND-GRL':[]}
    MaxLccList = {'HDA':[],'FINDER':[],'ND-GRL':[]}
else:
#     colors = {'HDA':'blueviolet','RatioCut':'blue','FINDER':'red'}
    markers = {'HDA':'s','RatioCut':'H','FINDER':'D'} 
    methods = ['HDA','RatioCut','FINDER']
    sol = {'HDA':[],'RatioCut':[],'FINDER':[]}
    MaxLccList = {'HDA':[],'RatioCut':[],'FINDER':[]}
    

# for method in methods:
#     for line in open('../data/%s/%s.txt'%(prob_type, method),'r'):
#         sol[method].append(int(line))
#     for line in open('../data/%s/%s_MaxCCList.txt'%(prob_type, method),'r'):
#         MaxLccList[method].append(np.double(line))
sol['HDA'] = HDA_sol
sol['FINDER'] = FINDER_sol
sol['ND-GRL'] = SDNDRL_sol
MaxLccList['HDA'] = HDA_anc
MaxLccList['FINDER'] = FINDER_anc
MaxLccList['ND-GRL'] = SDNDRL_anc

target_attack = 0.2
threshold = sol.copy()# nodes set of those whose removal cause the target attack effect
for method in methods:
    lcc = MaxLccList[method]
    for i in range(len(lcc)):
        if lcc[i] < target_attack:
            threshold[method] = threshold[method][:i]
            break  
```

```python
def SubplotDataPartial(ax, prob_type, tag):
    for method in methods:
        y_score = MaxLccList[method]
        x_score = [0]
        current_weight = 0.0
        for node in sol[method]:
            current_weight += nodes_weight_d[str(node)]
            x_score.append(current_weight/total_nodes_weight)
#             current_weight += nodes_weight[int(node)]


        y_highlight = [0.74, 0.69, 0.64, 0.59, 0.54]
    
    
#         y_highlight = [0.84, 0.79, 0.74, 0.64, 0.54]
        
        
#         y_highlight = [0.84, 0.79, 0.74, 0.64, 0.54]

        
#         y_highlight = [0.84, 0.74, 0.64, 0.59, 0.54]


#         y_highlight = [0.89, 0.79, 0.69, 0.59, 0.54]

    
        locs = []
        for y in y_highlight:
            for i in range(len(y_score)):
                if y_score[i] <= y:
                    locs.append(i)
                    break
        x_highlight = [x_score[item] for item in locs]
        y_highlight = [y_score[item] for item in locs]
        ax.scatter(x=x_highlight, y=y_highlight, s=1800, c=colors[method], marker=markers[method])
        
        ax.plot(x_score, y_score, c=colors[method], marker=markers[method], linestyle='-',linewidth=9, markersize=5,label='%s'%method)

    ax.tick_params(axis='x', labelsize=TickSize)
    ax.tick_params(axis='y', labelsize=TickSize)
    
    xmajorLocator = MultipleLocator(0.10) #将x主刻度标签设置为0.05的倍数
    xmajorFormatter = FormatStrFormatter('%.2f') #设置x轴标签文本的格式
    ax.xaxis.set_major_locator(xmajorLocator)
    ax.xaxis.set_major_formatter(xmajorFormatter)
    
    ymajorLocator = MultipleLocator(0.05) #将y轴主刻度标签设置为的倍数
    ymajorFormatter = FormatStrFormatter('%.2f') #设置y轴标签文本的格式
    ax.yaxis.set_major_locator(ymajorLocator)
    ax.yaxis.set_major_formatter(ymajorFormatter)
    
    legend = ax.legend(loc='upper right',bbox_to_anchor=(1, 0.98), fontsize=LegendSize, frameon=True, framealpha=1.0)
    legend.get_frame().set_linewidth(8.0)
    
    ax.spines['top'].set_linewidth(5)
    ax.spines['right'].set_linewidth(5)
    ax.spines['bottom'].set_linewidth(5)
    ax.spines['left'].set_linewidth(5)
    
    
    ax.set_xlim([0,0.5])
    ax.set_ylim([0.49,0.8])
    
    
#     ax.set_xlim([0,0.5])
#     ax.set_ylim([0.49,0.95])
    
    
#     ax.set_xlim([0,0.7])
#     ax.set_ylim([0.49,1.0])
    
    
#     ax.set_xlim([0,0.5])
#     ax.set_ylim([0.49,0.9])
    
    
#     ax.set_xlim([0,0.7])
#     ax.set_ylim([0.49,1.0])
    
    
    ax.set_xlabel('Fraction of Removed Nodes\' Costs(Weights)', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
    ax.set_ylabel('Residual Cycle Node Ratio(CNR)', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
        
    ax.text(-0.18, 1.03, '%s'%tag, size=subplotTitleSize+10, transform=ax.transAxes, weight='bold')
    
    
    ax.set_title('Yeast - degree-weighted - CNR Curve', pad = Labelpad, fontdict={'fontsize': subplotTitleSize, 'fontweight' : 'bold', 'verticalalignment': 'baseline', 'horizontalalignment': 'center'})
#     ax.set_title('NS_GC - degree-weighted - CNR Curve', pad = Labelpad, fontdict={'fontsize': subplotTitleSize, 'fontweight' : 'bold', 'verticalalignment': 'baseline', 'horizontalalignment': 'center'})
#     ax.set_title('Jazz - degree-weighted - CNR Curve', pad = Labelpad, fontdict={'fontsize': subplotTitleSize, 'fontweight' : 'bold', 'verticalalignment': 'baseline', 'horizontalalignment': 'center'})
#     ax.set_title('Email - degree-weighted - CNR Curve', pad = Labelpad, fontdict={'fontsize': subplotTitleSize, 'fontweight' : 'bold', 'verticalalignment': 'baseline', 'horizontalalignment': 'center'})
#     ax.set_title('C.elegans - degree-weighted - CNR Curve', pad = Labelpad, fontdict={'fontsize': subplotTitleSize, 'fontweight' : 'bold', 'verticalalignment': 'baseline', 'horizontalalignment': 'center'})

    
def Histogram(Ax, method, tag):
    Ax.axis('tight')
    Ax.spines['top'].set_color('none')
    Ax.spines['right'].set_color('none')
    
    Ax.spines['bottom'].set_linewidth(5)
    Ax.spines['left'].set_linewidth(5)

    values = [degrees[str(node)] for node in threshold[method]]
    
    min_val = np.min(values)
    max_val = np.max(values)
    Ax.hist(values,bins=10,density=True, color=colors[method], edgecolor='w', linewidth=3.2)
    sns.kdeplot(values,fill=True, ax=Ax, linewidth=3.2)
    
    Ax.tick_params(axis='x', labelsize=TickSize)
    Ax.tick_params(axis='y', labelsize=TickSize)
    
    
    
#     Ax.set_xlim([0,120])
#     Ax.set_ylim([0,0.08])
    if method == 'HDA':
        Ax.set_xlim([0,120])
        Ax.set_ylim([0,0.05])
        Ax.set_xticks(np.arange(0, 120, 20))
    elif method == 'FINDER':
        Ax.set_xlim([0,120])
        Ax.set_ylim([0,0.05])
        Ax.set_xticks(np.arange(0, 120, 20))
    elif method == 'ND-GRL':
        Ax.set_xlim([0,70])
        Ax.set_ylim([0,0.07])
        Ax.set_xticks(np.arange(0, 70, 20))
    
#     Ax.set_xlim([0,35])
#     Ax.set_ylim([0,0.12])
#     if method == 'HDA':
#         Ax.set_xlim([0,45])
#         Ax.set_ylim([0,0.14])
#         Ax.set_xticks(np.arange(0, 45, 20))
#     elif method == 'FINDER':
#         Ax.set_xlim([0,45])
#         Ax.set_ylim([0,0.14])
#         Ax.set_xticks(np.arange(0, 45, 20))
#     elif method == 'ND-GRL':
#         Ax.set_xlim([0,45])
#         Ax.set_ylim([0,0.16])
#         Ax.set_xticks(np.arange(0, 45, 20))
    
    
#     Ax.set_xlim([0,100])
#     Ax.set_ylim([0,0.04])
#     if method == 'HDA':
#         Ax.set_xlim([0,120])
#         Ax.set_ylim([0,0.035])
#         Ax.set_xticks(np.arange(0, 120, 20))
#     elif method == 'FINDER':
#         Ax.set_xlim([0,120])
#         Ax.set_ylim([0,0.03])
#         Ax.set_xticks(np.arange(0, 120, 20))
#     elif method == 'ND-GRL':
#         Ax.set_xlim([0,120])
#         Ax.set_ylim([0,0.03])
#         Ax.set_xticks(np.arange(0, 120, 20))
    
    
#     Ax.set_xlim([0,65])
#     Ax.set_ylim([0,0.05])
#     if method == 'HDA':
#         Ax.set_xlim([0,80])
#         Ax.set_ylim([0,0.05])
#         Ax.set_xticks(np.arange(0, 80, 20))
#     elif method == 'FINDER':
#         Ax.set_xlim([0,80])
#         Ax.set_ylim([0,0.05])
#         Ax.set_xticks(np.arange(0, 80, 20))
#     elif method == 'ND-GRL':
#         Ax.set_xlim([0,80])
#         Ax.set_ylim([0,0.05])
#         Ax.set_xticks(np.arange(0, 80, 20))
    
    
#     Ax.set_xlim([0,120])
#     Ax.set_ylim([0,0.05])
#     if method == 'HDA':
#         Ax.set_xlim([0,100])
#         Ax.set_ylim([0,0.05])
#         Ax.set_xticks(np.arange(0, 100, 20))
#     elif method == 'FINDER':
#         Ax.set_xlim([0,100])
#         Ax.set_ylim([0,0.05])
#         Ax.set_xticks(np.arange(0, 100, 20))
#     elif method == 'ND-GRL':
#         Ax.set_xlim([0,100])
#         Ax.set_ylim([0,0.05])
#         Ax.set_xticks(np.arange(0, 100, 20))


    Ax.set_xlabel('Degree of the Removed Nodes in Yeast', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
#     Ax.set_xlabel('Degree of the Removed Nodes in NS_GC', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
#     Ax.set_xlabel('Degree of the Removed Nodes in Jazz', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
#     Ax.set_xlabel('Degree of the Removed Nodes in Email', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
#     Ax.set_xlabel('Degree of the Removed Nodes in C.elegans', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
    
    Ax.text(-0.1, 1.03, '%s'%tag, size=subplotTitleSize+10, transform=Ax.transAxes, weight='bold')
    Ax.set_ylabel('Proportion', fontweight='bold',fontsize=TickSize+5, labelpad =Labelpad)
    Ax.set_title('%s'%(method), pad = Labelpad, fontdict={'fontsize': subplotTitleSize, 'fontweight' : 'bold', 'verticalalignment': 'baseline', 'horizontalalignment': 'center'})
```
